```python
import pandas as pd

# --- PHASE 1: THE RAW DATA ---
# This simulates the "Messy" export from a bank or internal system
raw_data = {
    'Transaction_ID': ['TXN_001', 'TXN_002', 'TXN_001', 'TXN_003', 'TXN_004', 'TXN_005'],
    'Amount': [50000, 1200, 50000, 450000, 10000, 75000],
    'Currency': ['KES', 'USD', 'KES', 'KES', 'USD', 'KES'],
    'Vendor': ['Microsoft', None, 'Microsoft', 'Unknown_Consultant', 'Amazon_AWS', 'Office_Supplies_Inc'],
    'Status': ['Paid', 'Pending', 'Paid', 'Urgent_Review', 'Paid', 'Paid']
}

df = pd.DataFrame(raw_data)

# --- PHASE 2: THE CONVERSION ENGINE ---
rate = 130
df['Amount_USD'] = df.apply(lambda x: round(x['Amount'] / rate, 2) if x['Currency'] == 'KES' else x['Amount'], axis=1)

# --- PHASE 3: THE FORENSIC FLAGS ---
# 1. Spot duplicates (Double Payments)
df['Duplicate_Risk'] = df.duplicated(subset=['Transaction_ID'], keep=False)

# 2. Spot Compliance Gaps (Missing Vendors)
df['KYC_Gap'] = df['Vendor'].isna()

# --- PHASE 4: THE AUDIT REPORT ---
# We filter for ONLY the rows that need a Senior Auditor's attention
audit_report = df[(df['Duplicate_Risk'] == True) | (df['KYC_Gap'] == True) | (df['Status'] == 'Urgent_Review')]

print("🚀 FORENSIC AUDIT COMPLETE")
print("-" * 30)
print(audit_report[['Transaction_ID', 'Amount_USD', 'Vendor', 'Status']])
```
